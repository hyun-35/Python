## 데이터 추가

```python
# '팔굽혀펴기' 열 추가

df['팔굽혀펴기'] = [15,12,10,4,20,24,9,11,11,17]
df
```

```python
# 새로운 행 추가

new_row = pd.Series([11,3,42.4,30,15,11,12], index = df.columns, name = 10)
new_row
```
